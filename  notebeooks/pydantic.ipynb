```python
from pydantic import BaseModel
from typing import List, Optional


class Student(BaseModel):
    usn:int
    name:str
    subject:List['str']
    isPresnt:Optional[bool]='true'
```

```python
s1 = Student(usn=123,name="Husen Basha",subject=["Python","Java","C++"],)
s1
```

```python
class Dept(BaseModel):
    cid:int
    cname:str
    student:Student

d1 = Dept(cid=452,cname='MCA',student=Student(usn=123,name="Husen Basha",subject=["Python","Java","C++"]))
d1
```
