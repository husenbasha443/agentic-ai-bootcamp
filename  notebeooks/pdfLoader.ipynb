```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("D:/agentic-ai-bootcamp/documents/RohanSQL.pdf")
final_load = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=100)
final_split = splitter.split_documents(final_load)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
texts = [doc.page_content for doc in final_split]
vector = embeddings.embed_documents(texts)
```

```python
from langchain_chroma import Chroma

store = Chroma.from_documents(documents=final_split,embedding=embeddings,persist_directory='D:/agentic-ai-bootcamp/documents/husendb')
```

```python
store.similarity_search("What is Database")
```
