### Load ENV and Model

```python
from dotenv import load_dotenv
from ollama import embeddings

load_dotenv()
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.1-8b-instant")
```

```python
from langchain_core.messages import HumanMessage,AIMessage
from pprint import pprint

result =model.invoke(
    [
        AIMessage(content="Hello How Can I Help You...",name="LLM"),
        HumanMessage(content="I want to learn coding",name="Husen"),
        AIMessage(content="Which Programming you prefer to learn",name="LLM"),
        HumanMessage(content="I Prefer Python",name="Husen")
    ]
)
result
```
